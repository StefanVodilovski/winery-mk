```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
# Initialize the webdriver.
browser = webdriver.Chrome(options=options)
```

```python
browser.get('https://wine.mk/shop/?filter_country=macedonia')
```

```python
print(browser.page_source)
```

```python

element = browser.find_elements(By.TAG_NAME,'del')
for i in element:
  browser.execute_script("""
  var element = arguments[0];
  element.parentNode.removeChild(element);
  """, i)

title_elements = browser.find_elements(By.CLASS_NAME,"woocommerce-loop-product__title")
price_elements = browser.find_elements(By.CLASS_NAME,"woocommerce-Price-amount")
image_elements = browser.find_elements(By.CLASS_NAME, "attachment-woocommerce_thumbnail")
matrix = []
for i in range(0,len(price_elements)):
    item = dict()
    item["Title"] = title_elements[i].text
    item["Price"] = price_elements[i].text
    item["Image"] = image_elements[i].get_attribute("src")
    matrix.append(item)

df = pd.DataFrame(matrix)

df
```

```python
df.to_csv("/content/data_for_wine/wines.csv",index=False)
```
