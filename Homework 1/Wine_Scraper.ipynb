```python
!apt update
!apt install chromium-chromedriver
!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
browser = webdriver.Chrome(options=options)
browser.get('https://wine.mk/product-category/wines/?filter_country=macedonia')
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

products = browser.find_element(By.CLASS_NAME, 'products').find_elements(By.CLASS_NAME, 'classic')
page_numbers = browser.find_elements(By.CLASS_NAME, 'page-numbers')

product = []
for i in range(0,len(products)):
  item = {}

  products2 = browser.find_element(By.CLASS_NAME, 'products').find_elements(By.CLASS_NAME, 'classic')

  title = products2[i].find_element(By.CLASS_NAME, 'woocommerce-loop-product__title').text
  item['Title'] = title

  price = products2[i].find_element(By.CLASS_NAME, 'price').text
  item['Price'] = price

  img_url = products2[i].find_element(By.TAG_NAME, 'img').get_attribute('src')
  item['Img Url'] = img_url

  links = products2[i].find_elements(By.TAG_NAME, 'a')
  links[2].click()
  additional = browser.find_element(By.CLASS_NAME, 'tabs').find_element(By.CLASS_NAME, 'additional_information_tab')
  additional.click()
  labels = browser.find_elements(By.CLASS_NAME, 'woocommerce-product-attributes-item__label')
  winery = browser.find_elements(By.CLASS_NAME, 'woocommerce-product-attributes-item__value')
  for i in range (0,len(labels)):
    if labels[i].text == "Macedonian Wineries:":
       winery = winery[i].text
       item['Winery'] = winery
  browser.back()


  product.append(item)

# df = pd.DataFrame(product)
# df
```

```python
page_numbers = browser.find_elements(By.CLASS_NAME, 'page-numbers')
for i in range(0,len(page_numbers)):
  page_numbers = browser.find_elements(By.CLASS_NAME, 'page-numbers')
  href = page_numbers[i].get_attribute('href')
  if href is not None:
    products = browser.find_element(By.CLASS_NAME, 'products').find_elements(By.CLASS_NAME, 'classic')

    for i in range(0,len(products)):
      item = {}

      products2 = browser.find_element(By.CLASS_NAME, 'products').find_elements(By.CLASS_NAME, 'classic')

      title = products2[i].find_element(By.CLASS_NAME, 'woocommerce-loop-product__title').text
      item['Title'] = title

      price = products2[i].find_element(By.CLASS_NAME, 'price').text
      item['Price'] = price

      img_url = products2[i].find_element(By.TAG_NAME, 'img').get_attribute('src')
      item['Img Url'] = img_url

      links = products2[i].find_elements(By.TAG_NAME, 'a')
      links[2].click()
      additional = browser.find_element(By.CLASS_NAME, 'tabs').find_element(By.CLASS_NAME, 'additional_information_tab')
      additional.click()
      labels = browser.find_elements(By.CLASS_NAME, 'woocommerce-product-attributes-item__label')
      winery = browser.find_elements(By.CLASS_NAME, 'woocommerce-product-attributes-item__value')
      for i in range (0,len(labels)):
        if labels[i].text == "Macedonian Wineries:":
          winery = winery[i].text
          item['Winery'] = winery
      browser.back()


      product.append(item)

df = pd.DataFrame(product)
df.to_csv('wine.csv', index=False)
df
```
