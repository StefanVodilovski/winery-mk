```python
# Install chromium, its driver, and selenium.
!apt update
!apt install chromium-chromedriver
!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
# Initialize the webdriver.
```

```python
browser = webdriver.Chrome(options=options)
# Navigate to a web page.
browser.get('https://wine.mk/product-category/wines/?filter_country=macedonia')
# Print the page source code.
```

```python
products = browser.find_element(By.CLASS_NAME, 'products').find_elements(By.CLASS_NAME, 'classic')
page_numbers = browser.find_elements(By.CLASS_NAME, 'page-numbers')

product = []
for i in range(0,len(products)):
  item = {}

  title = products[i].find_element(By.CLASS_NAME, 'woocommerce-loop-product__title').text
  item['Title'] = title

  price = products[i].find_element(By.CLASS_NAME, 'price').text
  item['Price'] = price

  img_url = products[i].find_element(By.TAG_NAME, 'img').get_attribute('src')
  item['Img Url'] = img_url

  product.append(item)

# df = pd.DataFrame(product)
# df
```

```python
page_numbers = browser.find_elements(By.CLASS_NAME, 'page-numbers')
for i in range(0,len(page_numbers)):
  page_numbers = browser.find_elements(By.CLASS_NAME, 'page-numbers')
  href = page_numbers[i].get_attribute('href')
  if href is not None:
    products = browser.find_element(By.CLASS_NAME, 'products').find_elements(By.CLASS_NAME, 'classic')

    for i in range(0,len(products)):
      item = {}

      title = products[i].find_element(By.CLASS_NAME, 'woocommerce-loop-product__title').text
      item['Title'] = title

      price = products[i].find_element(By.CLASS_NAME, 'price').text
      item['Price'] = price

      img_url = products[i].find_element(By.TAG_NAME, 'img').get_attribute('src')
      item['Img Url'] = img_url

      product.append(item)

df = pd.DataFrame(product)
df.to_csv('wine.csv', index=False)
df
```
